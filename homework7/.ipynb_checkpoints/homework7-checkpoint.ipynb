```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 1000)
amplitude = np.array([0.5, 1, 1.5, 2, 2.5])
frequency = np.array([1, 2, 3, 4, 5])
def make_sine_wave(x, A, w):
    y = A * np.sin(w * x)
    return y
```

```python
fig, axes = plt.subplots(1, 3, figsize=(25, 5))

for A in amplitude:
    y = make_sine_wave(x, A, 1)
    axes[0].plot(x, y)
axes[0].set_title("Varying Amplitude")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

for w in frequency:
    y = make_sine_wave(x, 1, w)
    axes[1].plot(x, y)
axes[1].set_title("Varying Frequency")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")

for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, w)
    axes[2].plot(x, y)
axes[2].set_title("Varying Freq. + Amp.")
axes[2].set_xlabel("x")
axes[2].set_ylabel("y")

# how should we write the legend for this problem???
plt.show()
```
