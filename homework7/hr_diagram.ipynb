```python
magnitude = data[:, 0]
temperature = data[:, 1]
print(magnitude)
print(np.shape(magnitude))
print(np.shape(temperature))
```

```python
# --- 3.3 Plot and HR Diagram ---
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)

plt.figure(figsize = (8, 6))
plt.scatter(log_temperature, log_magnitude)

brightest = np.argmin(magnitude)
dimmest = np.argmax(magnitude)
hottest = np.argmax(temperature)
coldest = np.argmin(temperature)

bright = plt.scatter(log_temperature[brightest], log_magnitude[brightest], color = "darkorchid", marker = '*', s=250)
dim = plt.scatter(log_temperature[dimmest], log_magnitude[dimmest], color = "orchid", marker = '*', s=250)
hot = plt.scatter(log_temperature[hottest], log_magnitude[hottest], color = "red", marker = '*', s=250)
cold = plt.scatter(log_temperature[coldest], log_magnitude[coldest], color = "paleturquoise", marker = '*', s=250)

plt.legend(title = 'Stars', handles = [bright, dim, hot, cold], labels = ['Brightest Star', 'Dimmest Star', 'Hottest Star', 'Coldest Star'])

plt.title("Hertzprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.show()
```
