```python
# --- 4: Prime Star Systems ---
arr = np.array([
    [2, 3, 5],
    [14, 6, 8],
    [11, 13, 17],
    [7, 10, 13]
])
```

```python
def is_prime(x):
    if x < 2:
        return False
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            return False
    return True

def prime_rows(arr):
    prime_mask = np.array([[is_prime(x) for x in row] for row in arr])
    return arr[prime_mask.any(axis=1)]

print(prime_rows(arr))
```

```python
# --- 6: The Expanding Universe ---
def spacing(string_num, spaces):
    string_num = np.array(string_num)
    result = []
    for x in string_num:
        new_string = ''
        for i in range(len(x)):
            new_string += x[i]
            if i != len(x) - 1:
                new_string += ' ' * spaces
        result.append(new_string)
    return np.array(result)

universe = np.array(['galaxy', 'clusters'])
print(spacing(universe, 1))
print(spacing(universe, 2))
```

```python
# --- 7: Second-Brightest Star ---
np.random.seed(123)
stars = np.random.randint(500, 2000, (5,5))

def secondDimmest(arr):
    result = []

    for row in arr:
        max_val = -np.inf
        second_max = -np.inf
        for val in row:
            if val > max_val: 
                second_max = max_val
                max_val = val
            elif val > second_max:
                second_max = val
        result.append(second_max)
    return np.array(result)

print(stars)
print(secondDimmest(stars))
```
